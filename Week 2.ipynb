```python
#pairplot

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\abdussamad p\Desktop\cancer dataset.csv")
sns.pairplot(df, vars=["radius_mean","texture_mean","perimeter_mean","area_mean","diagnosis"])
plt.show()

#Findings:
#area_mean, radius_mean and perimeter_mean values are more for malignant patients and less for for benign patients.
```
